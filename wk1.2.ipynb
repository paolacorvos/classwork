- strings
- int

- boolean

## Complex Variables

- list

```python
# string
string = "any sequence of characters" # strings use quotes or ''

# never name variables this crappy way, like 's'
s = str("something else")

# use underscore for naming variables, not hyphens, not NumberString, not numberString
number_string = "123"

print(type(string))
print(type(s),s)

# Jupyter will print (display) the last line
number_string
```

```python
# int (integer)
x = 1
y = int('3')

print(x)
print(type(y))

# aligning info to output
print("x plus y equals",x+y, ", but no f string")
print(f"x plus y equals {x+y}, but with an f string")

# concatination is different from addition
string + str(x)
```

```python
# float

# we can overwrite variable names
x = 1.0
y = float(42)

print(f"x = {x}, y = {y}, and y is a {type(y)}")
print(f"x + y = {x+y}, but y + 5 = {y+5}")
```

```python
# boolean = true/false

is_true = True # False
print(is_true)

is_false = bool(0)
print(is_false)
```

```python
# variables store the resolution of code
outcome = x + y
print(outcome)
```

```python
# floats and ints have a few methods
x = 10
y = 5

# +,-,*,/,%, 

print(x*y)
print(x/y)
print(x%y, "modulus")
print(x/3, "normal division")
print(x//3, "floor division")
```
